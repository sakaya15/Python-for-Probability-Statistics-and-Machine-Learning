```python
#10ポイントと20ポイントのみ表が出るとポイントが発生する時の、合計の期待値E(ξ|η)を求める
#h=aη+bを用いた解法

from sympy.abc import a,b#変数a,bを指定
X10,X20,X50 = S.symbols("X10,X20,X50",real=True)#変数X10,X20,X50を指定
xi  = 10*X10+20*X20+50*X50#ξを定義
eta = X10*X20*30 + X10*(1-X20)*(10)+ (1-X10)*X20*(20)#ηを定義
h = a*eta + b#h関数を定義
MSE=S.summation((xi-h)**2*S.Rational(1,8),(X10,0,1),(X20,0,1),(X50,0,1))#MSEを定義、summation:総和,Rational:分数
sol=S.solve([S.diff(MSE,a),S.diff(MSE,b)],(a,b))#solve:方程式を解く、diff:微分
print (sol)
```

```python
#<ξ-h,η>=0,<ξ-h,1>=0から導く解法
#<ξ-h,η>=0を展開

expr=S.expand((xi-h)*eta)#expand:展開
print(expr)
```

```python
expr.xreplace({X10**2:0.5, X20**2:0.5,X10:0.5,X20:0.5,X50:0.5})#xreplace:X10**2、X20**2、X10、X20、X50をそれぞれ0.5に代入
```

```python
#<ξ-h,1>=0を展開


print (S.expand((xi-h)*1).xreplace({X10**2:0.5, X20**2:0.5,X10:0.5,X20:0.5,X50:0.5}))#expand:展開,xreplace:それぞれ0.5を代入
```
